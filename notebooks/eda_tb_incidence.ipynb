# **Tuberculosis Incidence Analysis (2000-2023)**
### World Bank World Development Indicators (WDI)

---

## **Dataset Metadata**
| Attribute | Value |
|-----------|-------|
| **Indicator** | Incidence of tuberculosis (per 100,000 people) |
| **Time Frame** | 2000-2023 |
| **Coverage** | 209 Economies |
| **Frequency** | Annual |
| **Last Updated** | 5 months ago |
| **Source** | [Global Tuberculosis Report, WHO](https://www.who.int/teams/global-tuberculosis-programme/data) |
| **Dataset** | World Development Indicators (WDI) |

**Definition**:  
Incidence of tuberculosis is the estimated number of new and relapse tuberculosis cases arising in a given year, expressed as the rate per 100,000 population. All forms of TB are included, including cases in people living with HIV.

## Import Libs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Load Data

```python
# Loads the file
df = pd.read_csv("/content/WB_WDI_SH_TBS_INCD_WIDEF.csv")
df.head()
```

## EDA

```python
df.info()
```

```python
# Shape and columns
print("Rows, Columns:", df.shape)
print("\nColumn Names:\n", df.columns.tolist())
```

```python
# Quick summary
df.describe(include='all')
```

### Keep Only Relevant Columns
Lets reduce the dataframe to just:
- REF_AREA_NAME → country or region name
- 2000 to 2023 → annual TB incidence

```python
# Keep region name + all years
columns_to_keep = ['REF_AREA_NAME'] + [str(year) for year in range(2000, 2024)]
df_tb = df[columns_to_keep].copy()
df_tb = df_tb.rename(columns={'REF_AREA_NAME': 'Region'})
df_tb.head()
```

### Convert Wide → Long Format
- Each `row` should be a `unique` (Region, Year) with `one TB incidence value`.

```python
df_long = df_tb.melt(id_vars='Region',
                     var_name='Year',
                     value_name='TB_Incidence')
df_long['Year'] = df_long['Year'].astype(int)
df_long = df_long.dropna(subset=['TB_Incidence'])
df_long.head()
```

### Dataset Summary

```python
print("Shape:", df_long.shape)
print("\nYears:", df_long['Year'].min(), "-", df_long['Year'].max())
print("\nUnique Regions:", df_long['Region'].nunique())
df_long.describe()
```

### Visualizing Global Trends

#### Average Global TB Trend

```python
# Set style and palette
sns.set(style="whitegrid")
sns.set_palette("colorblind")

# Create plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_long, x='Year', y='TB_Incidence',
             estimator='mean', ci='sd', lw=2.5, marker='o')

# Add titles and labels
plt.title("Global Trend of Tuberculosis Incidence (2000–2023)", fontsize=16, weight='bold')
plt.suptitle("Incidence per 100,000 people | Shaded area = Standard Deviation", fontsize=10)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Mean TB Incidence", fontsize=12)

# Removes top and right spines
sns.despine()

# Grid and layout
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

#### Individual Country Examples

```python
selected = ['Pakistan', 'India', 'China', 'South Africa', 'United States']
df_sample = df_long[df_long['Region'].isin(selected)]

plt.figure(figsize=(14, 7))
sns.lineplot(data=df_sample, x='Year', y='TB_Incidence', hue='Region', marker='o', lw=2.5)

plt.title("Tuberculosis Incidence in Selected Countries (2000–2023)", fontsize=16, weight='bold')
plt.ylabel("Incidence per 100,000", fontsize=12)
plt.xlabel("Year", fontsize=12)
plt.legend(title='Country', fontsize=10, title_fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)
sns.despine()
plt.tight_layout()
plt.show()
```

#### Heatmap of TB Incidence by Region and Year

```python
plt.figure(figsize=(16, 24))
df_pivot = df_long.pivot(index='Region', columns='Year', values='TB_Incidence')

sns.heatmap(df_pivot, cmap='mako', linewidths=0.05, linecolor='gray', cbar_kws={'label': 'TB Incidence'})

plt.title("Heatmap: Tuberculosis Incidence per 100,000 People (by Region & Year)", fontsize=16, weight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Region", fontsize=12)
plt.tight_layout()
plt.show()
```

#### Highest and Lowest in 2023

```python
df_2023 = df_long[df_long['Year'] == 2023]
top_10 = df_2023.sort_values(by='TB_Incidence', ascending=False).head(10)
bottom_10 = df_2023.sort_values(by='TB_Incidence', ascending=True).head(10)

print(" Highest TB Incidence in 2023:")
print(top_10[['Region', 'TB_Incidence']])

print("\n Lowest TB Incidence in 2023:")
print(bottom_10[['Region', 'TB_Incidence']])
```

```python
# Sort data
df_2023_sorted = df_2023.sort_values(by="TB_Incidence", ascending=False)

# Global style
sns.set(style="whitegrid", palette="colorblind")

# Set font styles globally
plt.rcParams.update({
    "font.family": "DejaVu Sans",         # Or "Arial", "Lato", "Calibri"
    "axes.titlesize": 20,
    "axes.titleweight": "bold",
    "axes.labelsize": 14,
    "axes.labelweight": "medium",
    "xtick.labelsize": 12,
    "ytick.labelsize": 11,
    "legend.fontsize": 12
})

# Calculate dynamic height
fig_height = len(df_2023_sorted) * 0.4
fig, ax = plt.subplots(figsize=(20, fig_height))

# Barplot
barplot = sns.barplot(
    data=df_2023_sorted,
    x="TB_Incidence",
    y="Region",
    hue="Region",
    dodge=False,
    legend=False,
    palette="viridis"
)

# Adjusts bar thickness
for bar in barplot.patches:
    bar.set_height(0.6)

# Add value labels
for index, value in enumerate(df_2023_sorted["TB_Incidence"]):
    ax.text(value + 1, index, f'{value:.1f}', va='center', fontsize=10)

# Titles and labels
ax.set_title("Tuberculosis Incidence per 100,000 People by Country (2023)")
ax.set_xlabel("Incidence Rate (per 100,000 people)")
ax.set_ylabel("Country")

# Style tweaks
ax.grid(axis='x', linestyle='--', alpha=0.5)
sns.despine()
plt.tight_layout()
plt.show()
```

## Animated Choropleth Map

```python
import pandas as pd
import plotly.express as px

# Melt your DataFrame from wide to long format
df_melted = df.melt(
    id_vars=["REF_AREA_NAME"],
    value_vars=[str(year) for year in range(2000, 2024)],
    var_name="Year",
    value_name="TB_Incidence"
)

# Renames for clarity
df_melted.rename(columns={"REF_AREA_NAME": "Country"}, inplace=True)

# Ensures correct types
df_melted["Year"] = df_melted["Year"].astype(int)
df_melted["TB_Incidence"] = pd.to_numeric(df_melted["TB_Incidence"], errors='coerce')

# Plotly animated choropleth
fig = px.choropleth(
    df_melted,
    locations="Country",
    locationmode="country names",
    color="TB_Incidence",
    hover_name="Country",
    animation_frame="Year",
    color_continuous_scale= 'Cividis',  # if not visible to you please pass one of these as string Cividis, YlOrRd, YlGnBu, Blues, Greens
    range_color=[0, df_melted["TB_Incidence"].max()],
    title="Tuberculosis Incidence per 100,000 People (2000–2023)"
)


# Enlarges the map and styling
fig.update_layout(
    width=1200,  # Increase width
    height=700,  # Increase height
    geo=dict(showframe=False, showcoastlines=True, projection_type='equirectangular'),
    title_font_size=24,
    font=dict(size=14),
    coloraxis_colorbar=dict(title="Incidence Rate", tickfont=dict(size=12), titlefont=dict(size=14)),
    margin=dict(l=20, r=20, t=60, b=20)
)

# Show the figure
fig.show()
```

## Plotly Scatter Plot

```python
import plotly.express as px
import pandas as pd

# Reshape the data
df_scatter = df.melt(
    id_vars=["REF_AREA_NAME"],
    value_vars=[str(year) for year in range(2000, 2024)],
    var_name="Year",
    value_name="TB_Incidence"
)

df_scatter["Year"] = df_scatter["Year"].astype(int)

#  Drop rows with missing or zero incidence
df_scatter = df_scatter[df_scatter["TB_Incidence"].notna()]
df_scatter = df_scatter[df_scatter["TB_Incidence"] > 0]

# Create scatter plot
fig = px.scatter(
    df_scatter,
    x="Year",
    y="TB_Incidence",
    color="REF_AREA_NAME",
    hover_name="REF_AREA_NAME",
    title="TB Incidence Over Time by Country and Region",
    labels={"TB_Incidence": "Incidence per 100,000 people"},
    height=600
)

# Styling
fig.update_traces(marker=dict(size=7, opacity=0.7, line=dict(width=0.5, color='black')))
fig.update_layout(
    plot_bgcolor="white",
    xaxis_title="Year",
    yaxis_title="TB Incidence",
    font=dict(size=14),
    legend_title="Region"
)

fig.show()
```
